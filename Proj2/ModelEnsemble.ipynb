```python
import numpy as np
import pandas as pd

nb = pd.read_csv('submission.csv', dtype=int)
mlp1 = pd.read_csv('submission-mlp.csv', dtype=int)
mlp2 = pd.read_csv('submission-mlp_v2.csv', dtype=int)
log = pd.read_csv('submission_logitistic.csv', dtype=int)
```

```python
data = pd.DataFrame()
data['id'] = nb['id'].copy()
data['nb'] = nb['pred'].copy()
data['log'] = log['pred'].copy()
data['mlp1'] = mlp1['pred'].copy()
data['mlp2'] = mlp2['pred'].copy()

newdata = pd.DataFrame(data['id'], columns=['id'], index=None)

newdata['pred'] = data[['nb', 'log', 'mlp1', 'mlp2']].mode(axis=1).min(axis=1).astype(int)

with pd.option_context('display.max_rows', -1, 'display.max_columns', 5):
    print(data)
newdata.to_csv('submission-ensemble.csv', index=False)
```
